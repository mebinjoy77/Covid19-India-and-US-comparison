```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('owid-covid-data.csv')
```

```python
data.head()
```

```python
data.fillna(0,inplace=True)
```

```python
data.head()
```

```python
s = data[data['location'] == 'India']
s['date']
u =  data[data['location'] == 'United States']
u
```

```python
x = s['date']
y = s['total_cases']
plt.plot(x,y)
x2 = u['date']
y2 = u['total_cases']
dt = u['date'].tolist()
plt.plot(x2,y2)
plt.title('Covid 19 Cases Comparison India and US')
plt.xlabel('Months')
plt.ylabel('No of Covid Cases in Thousands')
plt.legend(['India','US'])
plt.xticks(x,dt)
plt.savefig('IndUSCovid.jpg')

```

This is the comparison of Covid Cases in India and the US

Now let's look at the average cases in these both countries

```python
print('US Average Cases per day -  ',u['new_cases'].mean())
print('India Average Cases per day - ',s['new_cases'].mean())
```

Now let's look at the maximum cases that was recorded in both countries

```python
print('US Maximum Cases reported per day -  ',u['new_cases'].max())
print('India Maximum Cases reported per day - ',s['new_cases'].max())
```

Note that this data is as of until 05/11/2020
